```python
from simstats import *
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
population = np.random.uniform(low=0, high=1000,size=1000000)
```

```python
population
```

```python
population.mean()
```

```python
population.std()
```

```python
sns.distplot(population)
```

```python
sim = simulator(population=population,effect=100, alpha=0.05, n=50)
sim.ci(sample=sim.xbase, alpha=0.5)
sim.ciPlot(base={'name': 'base', 'color': 'green'}, variant={'name': 'variant', 'color': 'blue'})
sim.ttestPlot()
sim.table()
```
