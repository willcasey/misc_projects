```python
import pandas as pd
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('/Users/wcasey/Desktop/ff_data.csv')
```

```python
df.plot(kind='scatter', x='PF', y= 'Win %')
```

```python
df.sort_values('Win %' ,ascending=False)
```

```python
df_points_per_year = df.groupby('year').agg({'PF': [np.mean, np.median, np.max, np.min, np.std]})['PF']
```

```python
df_points_per_year['year'] = df_points_per_year.index
```

```python
df_points_per_year
```

```python
df_points_per_year.drop(['std', 'year'], axis=1).plot(kind='line')
```

```python
df = df.merge(df_points_per_year, left_on = 'year', right_on = 'year')
```

```python
df['total_mean'] = df.PF.mean()
df['total_std'] = df.PF.std()
```

```python
df
```

```python
df['PF-Med'] = df.PF - df['median']
df['PF-Mean'] = (df.PF - df['mean'])/df['std']
```

```python
dfSort = df.sort_values('PF-Mean', ascending=False).reset_index(drop=True)
```
